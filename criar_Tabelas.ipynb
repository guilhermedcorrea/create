```python
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, DoubleType, TimestampType
import csv

def generate_schemas(file_name):
    schemas = {}

    with open(file_name, mode='r', encoding="utf-8") as csv_file:
        csv_reader = csv.DictReader(csv_file, delimiter=';', quotechar='"')

        for row in csv_reader:
            table_name = row['Tabela'].lower()
            if table_name not in schemas:
                schemas[table_name] = []

            column_name = row['estrutura']
            data_type = StringType()  
           
            if 'Tipo' in row:
                if row['Tipo'].lower() == 'int':
                    data_type = IntegerType()
                elif row['Tipo'].lower() == 'double':
                    data_type = DoubleType()
                elif row['Tipo'].lower() == 'timestamp':
                    data_type = TimestampType()

            schemas[table_name].append((column_name, data_type, True))

    for table_name, fields in schemas.items():
        schemas[table_name] = StructType([StructField(col_name, data_type, nullable) for col_name, data_type, nullable in fields])

    return schemas

file_name = r"C:\mlflowjobs\tabelas_teste.csv"
schemas = generate_schemas(file_name)


for table_name, schema in schemas.items():
    print(f"Tabela: {table_name}")
    print("schema = StructType([ \\")
    for field in schema.fields:
        print(f"    StructField(\"{field.name}\",{field.dataType},{field.nullable}), \\")
    print("  ])\n")
```
